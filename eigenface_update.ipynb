```python
import numpy as np
import glob
from PIL import Image
```

```python
directory = "data/faces94/male/9326871"
```

```python
jpgs = glob.glob(directory + '/*.jpg')
```

```python
imgs = np.array([np.array(Image.open(img).convert('L')) for img in jpgs])
```

```python
mean_img = np.mean(imgs, 0)
```
